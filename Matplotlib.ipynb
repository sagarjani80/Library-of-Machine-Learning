```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
heart_disease = pd.read_csv("heart-disease.csv")
plt.style.use("seaborn-whitegrid")
over_50 = heart_disease[heart_disease['age'] > 50]
fig , (ax0 , ax1,ax2) = plt.subplots(figsize = (10,10),
                                nrows = 3,
                                ncols = 1)
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax0.legend(*scatter.legend_elements(),title = "target")
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");
ax0.set(title = "heart disease of human",
       xlabel = "age",
       ylabel = "cholestrol");
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"])
ax1.legend(*scatter.legend_elements(),title = "sagar")
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");
ax1.set(title = "heart disease of animal",
       xlabel = "age",
       ylabel = "thalach");
scatter = ax2.scatter(x = over_50["age"],
                     y = over_50["trestbps"],
                     c = over_50["target"] )
ax2.axhline(y = over_50["trestbps"].mean(),
           linestyle = "--");
ax2.legend(*scatter.legend_elements(),title = "target")
fig.suptitle("Heart Disease Anaylsis" , fontsize = 20 , fontweight = "bold");
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
heart_disease = pd.read_csv("heart-disease.csv")
plt.style.use("seaborn-whitegrid")
over_50 = heart_disease[heart_disease['age'] > 50]
fig , (ax0 , ax1,ax2) = plt.subplots(figsize = (10,10),
                                nrows = 3,
                                ncols = 1)
scatter = ax0.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                     cmap = "winter")
ax0.legend(*scatter.legend_elements(),title = "target")
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--");
ax0.set(title = "heart disease of human",
       xlabel = "age",
       ylabel = "cholestrol");
ax0.set_xlim([50,80])
scatter = ax1.scatter(x = over_50["age"],
                     y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "winter")
ax1.legend(*scatter.legend_elements(),title = "sagar")
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");
ax1.set(title = "heart disease of animal",
       xlabel = "age",
       ylabel = "thalach");
ax1.set_xlim([50,100])
scatter = ax2.scatter(x = over_50["age"],
                     y = over_50["trestbps"],
                     c = over_50["target"] ,
                     cmap = "winter")
ax2.axhline(y = over_50["trestbps"].mean(),
           linestyle = "--");
ax2.legend(*scatter.legend_elements(),title = "target")
fig.suptitle("Heart Disease Anaylsis" , fontsize = 20 , fontweight = "bold");
```
