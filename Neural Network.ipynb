### Preprocessing of image

```python
IMG_SIZE = 224
def process_image(image_path):
    # read the image
    image = tf.io.read_file(image_path)
    # turn jpeg image in to numerical tensor with 3 color channel (red,green,blue)
    image = tf.image.decode_jpeg(image,channels=3)
    #convert color channel value 0-255 to 0-1
    image = tf.image.convert_image_dtype(image,tf.float32)
    # resize the image
    image = tf.image.resize(image,size = [IMG_SIZE,IMG_SIZE])
    return image
process_image(filename[10])
```

### create function of return label

```python
def get_image_label(image_path,label):
    image = process_image(image_path)
    return image,label
get_image_label(x[10],y[10])
```

### Create a data batches
-> if we have 10,000 images and we want to process in one go. so they all might not fit in to memory. that's why we do 32(1 batch size) image at a time 

```python
# if we have vadidation dataset than we no need to shuffle them
# if we have training data than we need to shuffle them

BATCH_SIZE = 32
def create_data_batches(X,Y,batch_size = BATCH_SIZE,test_data = False,valid_data = False):
    if test_data:
        print("Create data batches for test data....")
        data = tf.data.Dataset.from_tenso(r_slices((tf.constant(X)))
        data_batch = data.map(process_image).batch(BATCH_SIZE)
        return data_batch
    elif valid_data:
        print("Create data batch for validation data....")
        data = tf.data.Dataset.from_tensor_slices((tf.constant(X),tf.constant(Y)))
        data_batch = data.map(get_image_label).batch(BATCH_SIZE)
        return data_batch
    else:
        print("Create data batch for training data....")
        data = tf.data.Dataset.from_tensor_slices((tf.constant(X) , tf.constant(Y)))
        data = data.shuffle(buffer_size = len(X))
        data_batch = data.map(get_image_label).batch(BATCH_SIZE)
        return data_batch
```

### Building a Modal
1) input shape
2) output shape
3) URL

```python
INPUT_SHAPE = [None,IMG_SIZE,IMG_SIZE,3]
OUTPUT_SHAPE = len(unique_breed)
URL = "https://tfhub.dev/google/imagenet/mobilenet_v2_130_224/classification/5"
```

```python
def create_model(input_shape = INPUT_SHAPE,output_shape = OUTPUT_SHAPE,model_url = URL):
    print("Building a model with : ", model_url)
    #setup a layers with sequential keras
    model = tf.keras.Sequential([
        hub.KerasLayer(model_url),
        tf.keras.layers.Dense(units = output_shape ,
                             activation = "softmax")
    ])
    
    #compile the model
    model.compile(
                    loss = tf.keras.losses.CategoricalCrossentropy(),
                    optimizer = tf.keras.optimizers.Adam(),
                    metrics = ["accuracy"]
                            )    
    #Build a model
    model.build(input_shape)
    
    return model
```

```python
model = create_model()
model.summary()
```

## Creating Callbacks
-> Callbacks is a helper function a model can use during training to do such things like save it progress , check it progress or stop training early
-> we will create two callbacks
    1) For TensorBoard which helps track our model progress 
    2) For early stoping which prevents our model from training for too long

```python
#load TensorBoard notebook extension
%load_ext tensorboard
```

```python
# Create a Tensor Board callbacks which is able to save log to a directory and pass it to our model's fit() function
import datetime
import os
def create_tensorboard_callback():
    logdir = os.path.join("Data/logs",
                         datetime.datetime.now().strftime("%y%m%d - %H%M%S"))
    return tf.keras.callbacks.TensorBoard(logdir)
```

 2) Create early stopping callbacks which helps model to stop overfitting when the accuracy metrix stops improving

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = "val_accuracy",
                                                 patience = 3)
```

## Train Model

```python
# epochs indicates number of passes of the entire training dataset the machine learning algorithm has completed
NUM_EPOCHS = 100
def train_model():
    #create model
    model = create_model()
    #crete tensorboard session every time we train a model
    tensorboard = create_tensorboard_callback()
    # Fit the model
    model.fit(x = train_data,
             epochs = NUM_EPOCHS,
             validation_data = val_data,
             validation_freq = 1,
             callbacks = [tensorboard,early_stopping])
    return model
```

```python
def get_pred_label(prediction_probability):
    return unique_breed[np.argmax(prediction_probability)]
```

## Unbatch the data
-> takes a batch dataset of(image,label) Tensors and return seprate arrays of images and labels

```python
def unbatchify(data):
    images = []
    labels = []
    
    for image,label in data.unbatch().as_numpy_iterator():
        images.append(image)
        labels.append(unique_breed[np.argmax(label)])
        
        return images,labels
    
val_images,val_labels = unbatchify(val_data)
```

## Visualize the prediction 

```python
def plot_pred(prediction_probability, labels, images, n=1):
    pred_prob,true_label,image = prediction_probability[n],labels[n],images[n]
    pred_label = get_pred_label(pred_prob)
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
    
    if(true_label == pred_label):
        color = "green"
    else:
        color = "red"
    plt.title("{} {:2.0f}% {}".format(pred_label,
                                     np.max(pred_prob)*100,
                                     true_label) , color = color)
```

```python
plot_pred(prediction_probability = predictions,labels = val_labels,images = val_images,n=0)
```

## Save a Model

```python
def save_model(model,suffix = None):
    modeldir = os.path.join("Data/Models",
                           datetime.datetime.now().strftime("%y%m%d-%H%M%S"))
    model_path = modeldir+"-"+suffix+".h5"
    model.save(model_path)
    return model_path
```

## Load a Model

```python
def load_model(model_path):
    model = tf.keras.models.load_model(model_path,
                                     custom_objects = {"KerasLayer" : hub.KerasLayer})
    return model
```
